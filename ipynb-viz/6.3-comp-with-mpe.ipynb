```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams.update({'font.size': 20})

categories = ['Unet:XL', 'O&F:baseline', 'BP:baseline']
datasets = ['comparing', 'Trios', 'RWC']
values_comparing = [83.0, 61.0, 40.0]
values_trios = [86.0, 71.2, 58.9]
values_rwc = [73.0, 61.6, 28.3]

x = np.arange(len(categories))
width = 0.25

# x = ['Unet:XL', 'O&F-MAESTRO', 'bp-pretrained']
# y = [83.0, 61.0, 40.0]

fig, ax = plt.subplots(figsize=(10, 10))

colours = ['lightgray','plum','lightblue','lightblue','palegreen', 'khaki']
# plt.bar(x, y, color=['lightgray','plum','lightblue','lightblue','palegreen', 'khaki'], yerr=[20, 10, 8], capsize=5)
bars_comparing = ax.bar(x-width, values_comparing, width, label='MuN,SWD,B10,PhA,CSD', color='tab:blue')
bars_trios = ax.bar(x, values_trios, width, label='TRI', color='tab:orange')
bars_rwc = ax.bar(x + width, values_rwc, width, label='RWC', color='tab:green')

ax.set_xlabel('Models')
ax.set_ylabel('Average Precision (%)')

ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```
