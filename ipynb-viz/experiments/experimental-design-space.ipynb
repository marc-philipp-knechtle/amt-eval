```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

matplotlib.rcParams.update({'font.size': 20})

# Data to plot
categories = ['Training size (files/hh:mm)', 'Training instrumentation (piano/nopiano)', 'Training diversity',
              'Architecture (OaF/BP)', 'Model size']
N = len(categories)
values = [1, 0, 0, 1, 1]
values_2 = [1, 0, 0, 0, 1]

# Create the radar chart
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()

# add first element such that the figure can be 'closed'
values += values[:1]
values_2 += values_2[:1]

angles += angles[:1]

fig, ax = plt.subplots(figsize=(12, 12), subplot_kw=dict(polar=True))

ax.fill(angles, values, color='red', alpha=0.25, label='Experiment A')
ax.plot(angles, values, color='red', linewidth=2)

ax.fill(angles, values_2, color='blue', alpha=0.25)
ax.plot(angles, values_2, color='blue', linewidth=2)

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

radial_levels = [1, 2]
ax.set_yticks(radial_levels)
radial_labels = ['Fixed', 'Variable']
ax.set_yticklabels(radial_labels)

# plt.title('Radar Chart', size=20, color='black', y=1.1)
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
import matplotlib

matplotlib.rcParams.update({'font.size': 20})

# Define data: 5 items × 10 categories with 4 categorical values (A, B, C, D)
categories = ['Data size (files/hh:mm)', 'Instrumentation (piano/nopiano)', 'Data diversity',
              'Architecture (OaF/BP)', 'Model size', 'Domain knowledge (seen/unseen)',
              'Model output (decoded/raw)']
items = ['Experiment A: Frame vs. NE', 'Experiment B: Training size', 'Experiment C: Training style', 'Experiment D: Training diversity', 'Experiment E: Model size']
category_values = ['Fixed', 'Varied during Test', 'Varied during Train', 'Varied during Train + Test']  # Categorical values

# Generate example categorical data
# np.random.seed(42)
# data_labels = np.random.choice(category_values, size=(1, 7))

data_numeric = [[2, 0, 2, 1, 0, 0, 1], # Experiment A: Comparing Frame output against Note events
                # Training size: Training, because we have different sizes in OaF vs BP baseline models
                # Architecture: Test, because we compare the two against each other
                # Test domain know: Test, because two different evals
                # Model output: Test, because that's the purpose of this exp
                [3, 1, 0, 1, 0, 1, 0], # Experiment B: Impact of the training size
                # Training size: train different models with diff size, eval with divverent size
                # Instrumentation: Eval piano - nopiano
                # Architecture: test - we train two models OaF + BP with this
                # Test domain: test
                [0, 3, 0, 1, 0, 1, 0], # Experiment C: Impact of the training style
                # Instrumentation: Training piano/nopiano models
                # Architecture: test OaF + BP
                # Test domain: test unseen + seen
                [0, 1, 3, 1, 0, 1, 0], # Experiment D: Impact of the training diversity
                #
                [0, 1, 0, 0, 3, 1, 1], # Experiment E: Impact of the model size
                ]


# Map categories to integers for colormap
# label_to_int = {label: i for i, label in enumerate(category_values)}
# data_numeric = np.vectorize(label_to_int.get)(data_labels)

# Define custom colors for the categories
# colors = ['lightgrey', '#20B2AA', '#FFE4B5', '#FF6347']
colors = ['lightgrey', '#20B2AA', 'lightsalmon', 'darkred']
cmap = ListedColormap(colors)

# Create the heatmap
fig, ax = plt.subplots(figsize=(25, 20))
im = ax.imshow(data_numeric, cmap=cmap)

# Set axis labels (this is necessary that the labels are on the correct positions
ax.set_xticks(np.arange(len(categories)))
ax.set_yticks(np.arange(len(items)))
# ax.grid(True, which='major', linestyle='-', linewidth='0.5')
# set shifted gridlabel annotations
ax.set_xticks(np.arange(len(categories)) + 0.5, minor=True)  # Shift x grid by 0.5
ax.set_yticks(np.arange(len(items)) + 0.5, minor=True)  # Shift y grid by 0.1
ax.grid(True, which='minor', linestyle='-', alpha=0.5)
ax.set_xticklabels(categories)
ax.set_yticklabels(items)

# Rotate category labels for readability
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Annotate each cell with the actual category label (A, B, C, D)
# for i in range(len(items)):
#     for j in range(len(categories)):
#         ax.text(j, i, data_labels[i, j], ha="center", va="center", color="black")

# Add a legend
from matplotlib.patches import Patch

legend_elements = [Patch(facecolor=colors[i], label=category_values[i]) for i in range(len(category_values))]
# previous loc was upper right
ax.legend(handles=legend_elements, title="Handling of the Design Space", bbox_to_anchor=(1.05, 1), loc='upper center')

# ax.set_title("Categorical Heatmap (4 Categories)")
plt.tight_layout()
plt.show()

```
